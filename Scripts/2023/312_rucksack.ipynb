```python
def commonFind(ilist):
    ordsList=[]
    for index,item in enumerate(ilist):
        halflen=int(len(item)//2) #finding the half length the easy way

        a=list(set(item[:halflen])&set(item[halflen:]))

        orda=[ord(char) - ord('a')+1 if char.islower() else ord(char)- ord('A')+27 for char in a]

        ordsList.append(sum(orda))
        
    return(sum(ordsList),ordsList)
```
